```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

# 控制点（你决定的点）
points = np.array([[0, 0], [1, 2], [3, 3], [4, 0], [5, 2]])
x, y = points[:, 0], points[:, 1]

# 插值B样条
t = np.linspace(0, 1, len(points))
spl = make_interp_spline(t, points, k=3)  # 三阶B样条

# 生成曲线点
t_new = np.linspace(0, 1, 200)
curve = spl(t_new)

plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置黑体支持中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号 '-' 显示问题
# 绘图
plt.plot(curve[:, 0], curve[:, 1], label='B-spline Curve')
plt.plot(x, y, 'o--', label='Control Points')
plt.legend()
plt.title("B-spline示意图")
plt.show()
```
